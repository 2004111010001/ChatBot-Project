```python
import pandas as pd
import numpy as np
```

**GIVING FLAG**

```python
df = pd.read_excel('HistoryLiveChat.xlsx')
```

```python
df=df[['ID','Chat content']];
```

```python
df.head()
```

```python
df = df.assign(Chat = df['Chat content'].str.split(r'\n')).explode('Chat')
```

```python
df.head(10)
```

```python
df['flag'] = np.where(df['Chat'].str.contains(r'Operator Badan Pusat (?!$)'), "A", "Q")
df.reset_index()
df.head()
```

**MEMISAHKAN TANGGAL DAN JAM**

```python
#df = df.assign(CleanChat = df['Chat'].apply(lambda x: pd.Series(str(x).split(":"))))
chat = df["Chat"]
new_chat = []
for x in chat:
    #print(x[20:])
    #print(x[5])
    try:
        if x[2] and x[5] == "-":
            new_chat.append(x[20:])
        else:
            new_chat.append(x)
    except IndexError:
        new_chat.append(x)
print(new_chat)
df = df.assign(CleanChat = new_chat)
df
    
```

**MEMISAHKAN USER**

```python
chat = df["CleanChat"]
new_chat = []
for x in chat:
    if ":" in x:
        new_chat.append(x[x.find(":")+2:])
        #print(x[x.find(":")+1:])
    else:
        new_chat.append(x)
df = df.assign(CleanChat2 = new_chat)
df
```

**MEMISAHKAN CHAT BERDASARKAN FLAG**

```python
df = pd.read_excel('CleanChatHistory.xlsx')
df=df[['ID','flag','CleanChat2']];
df
```

```python
question = []
indexQ = 0
flag = df["flag"]
for x in flag:
    if x == "Q":
        question.append(df.loc[indexQ, "CleanChat2"])
    else:
        question.append('')
    indexQ += 1

print(question)

answer = []
indexA = 0
flag = df["flag"]
for x in flag:
    if x == "A":
        answer.append(df.loc[indexA, "CleanChat2"])
    else:
        answer.append('')
    indexA += 1

print(answer)
df = df.assign(Question = question)
df = df.assign(Answer = answer)
df
```
